```python
# função com estatisticas sobre colunas com dados faltantes

def stats_dados_faltantes(df: pd.DataFrame) -> None:

  stats_dados_faltantes = []
  for col in df.columns:
    if df[col].isna().any():
      qtd, _ = df[df[col].isna()].shape
      total, _ = df.shape
      dict_dados_faltantes = {col: {'quantidade': qtd, "porcentagem": round(100 * qtd/total, 2)}}
      stats_dados_faltantes.append(dict_dados_faltantes)

  for stat in stats_dados_faltantes:
    print(stat)
```

```python
# Função para plotar os boxplot

def plot_boxplots(df, categorical_vars):
    """
    Plota boxplots para as variáveis categóricas em relação à variável contínua 'Exam_Score'.
    
    Parâmetros:
    df : DataFrame
        O DataFrame que contém os dados.
    categorical_vars : list
        Lista com os nomes das variáveis categóricas.
    """
    # Criar figura para os três gráficos lado a lado
    plt.figure(figsize=(18, 6))

    # Loop para criar um boxplot para cada variável categórica
    for i, var in enumerate(categorical_vars, 1):
        plt.subplot(1, len(categorical_vars), i)
        sns.boxplot(hue=var, y='Exam_Score', data=df, width=0.8, gap= 0.25)
        plt.ylim(50, 105)  # Limitar o eixo y de 0 a 110
        plt.title(f'Exam Score vs {var}')
        plt.xlabel(var)
        plt.ylabel('Exam Score')

    # Exibir os gráficos
    plt.tight_layout()
    plt.show()
```

```python
# Função grafico SCATTER

def plot_scatter(df, x_var, y_var):
    """
    Plota um gráfico de dispersão entre duas variáveis de um DataFrame.
    
    Parâmetros:
    df : DataFrame
        O DataFrame que contém os dados.
    x_var : str
        O nome da variável a ser plotada no eixo x.
    y_var : str
        O nome da variável a ser plotada no eixo y.
    """
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=x_var, y=y_var, alpha=0.6)
    plt.title(f'Relação entre {x_var} e {y_var}')
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.xlim(df[x_var].min() - 5, df[x_var].max() + 5)  # Ajustar limite do eixo x
    plt.ylim(45, 110)  # Ajustar limite do eixo y
    plt.grid()
    plt.show()
```

## 5\. Modelagem

#### Criação de features (dummies) 

```python
df_modelagem = df.copy()
```

```python
# preparando dados para utilizar o random forest

df_modelagem = pd.get_dummies(df_modelagem, dtype= float)
df_modelagem.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

import numpy as np
```

```python
# separando as variaveis explicativas e variavel resposta

X = df_modelagem.drop(['Exam_Score'], axis= 1)
y = df_modelagem['Exam_Score']
```

```python
# Separando dados em conjuntos de treino e teste

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 1991)
```

```python
# Criação do modelo

modelo = RandomForestRegressor(random_state= 1991)
```

```python
# Validação cruzada com 5 folds

cv_scores = cross_val_score(modelo, X_train, y_train, cv=5, scoring='neg_root_mean_squared_error')
```

```python
# Calcular a média e o desvio padrão dos scores
mean_cv_score = np.mean(cv_scores)
std_cv_score = np.std(cv_scores)

print(f'Mean CV Score (MSE): {-mean_cv_score}')
print(f'Standard Deviation of CV Scores: {std_cv_score}')
```

#### Usando RandomizedSearchCV

Randomized realiza uma busca aleatória em um conjunto de hiperparâmetros

```python
from sklearn.model_selection import RandomizedSearchCV

# Definir o espaço de busca para os hiperparâmetros

param_dist = {
    'n_estimators': np.arange(50, 300, 50),  # Número de árvores
    'max_features': [None, 'sqrt', 'log2'],  # Número de features consideradas
    'max_depth': [None, 10, 20, 30],  # Profundidade máxima da árvore
    'min_samples_split': [2, 5, 10],  # Mínimo de amostras para dividir um nó
    'min_samples_leaf': [1, 2, 4],  # Mínimo de amostras em cada folha
    'bootstrap': [True, False]  # Usar ou não amostras com reposição
}
```

```python
# Criar o RandomizedSearchCV

random_search = RandomizedSearchCV(
    modelo, param_distributions=param_dist, 
    n_iter=30,  # Número de combinações diferentes a serem testadas
    cv=5,  # Validação cruzada de 5 folds
    random_state=1991, 
    n_jobs=-1  # Usar todos os processadores disponíveis
)
```

```python
random_search.fit(X_train, y_train)
```

```python
# Ver os melhores hiperparâmetros encontrados

print("Melhores hiperparâmetros encontrados:")
print(random_search.best_params_)
```

```python
# O ** desempacota o dicionário dos melhores hiperparâmetros e os aplica no modelo diretamente.

modelo_otimizado = RandomForestRegressor(**random_search.best_params_)
```

```python
cv_scores_tunado = cross_val_score(modelo_otimizado, X_train, y_train, cv=5, scoring='neg_root_mean_squared_error')
```

```python
# Calcular a média e o desvio padrão dos scores
mean_cv_score_tunado = np.mean(cv_scores_tunado)
std_cv_score_tunado = np.std(cv_scores_tunado)

print(f'Mean CV Score (MSE): {-mean_cv_score_tunado}')
print(f'Standard Deviation of CV Scores: {std_cv_score_tunado}')
```

```python
# Treinar o modelo

modelo_otimizado.fit(X_train, y_train)
```

## 6\. Avaliando a qualidade do modelo


Calculando a média dos erros entre as previsões e o valores reais

```python
from sklearn.metrics import root_mean_squared_error
```

```python
y_pred = modelo_otimizado.predict(X_test)
```

```python
rmse = root_mean_squared_error(y_test, y_pred)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

Medindo a proporção da variância explicada pelo modelo em relação à variância total dos dados

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print(f"R² Score: {r2}")
```

#### Envolvimento dos pais

```python
# Comparar estatísticas descritivas em relação ao envolvimento dos pais

df.groupby('Parental_Involvement')['Exam_Score'].describe()
```

```python
from scipy.stats import ttest_ind

# Filtrar os dados para alto e baixo envolvimento dos pais
high_involvement = df[df['Parental_Involvement'] == 'High']['Exam_Score']
low_involvement = df[df['Parental_Involvement'] == 'Low']['Exam_Score']

# Realizar o teste t
t_stat, p_value = ttest_ind(high_involvement, low_involvement)

print(f'Estatística t: {t_stat}, Valor p: {p_value}')
```

O valor p foi menor que 0.05, podemos concluir que há uma diferença estatisticamente significativa entre os dois grupos.  
Apesar da diferença da média ser marginal entre Alto e Baixo envolvimento.
- Porém nenhum aluno com baixo envolvimento dos pais foi capaz de tirar nota 100.

#### Como Access_to_Resources - Disponibilidade de recursos educacionais (Baixa, Média, Alta) influenciam a nota

```python
# Comparar estatísticas descritivas por nível de acesso a recursos

df.groupby('Access_to_Resources')['Exam_Score'].describe()
```

```python
from scipy.stats import ttest_ind

# Filtrar os dados para alto e baixo envolvimento dos pais
high_involvement = df[df['Access_to_Resources'] == 'High']['Exam_Score']
low_involvement = df[df['Access_to_Resources'] == 'Low']['Exam_Score']

# Realizar o teste t
t_stat, p_value = ttest_ind(high_involvement, low_involvement)

print(f'Estatística t: {t_stat}, Valor p: {p_value}')
```

O valor p foi menor que 0.05, podemos concluir que há uma diferença estatisticamente significativa entre os dois grupos.  
Podemos concluir que a desigualdade no acesso aos recursos afeta diretamente a performance acadêmica
- Investir em infraestruturas educacionais que forneçam recursos adequados para todos os estudantes pode ser uma política eficiente para nivelar o desempenho entre alunos de diferentes contextos socioeconômicos, proporcionando a todos as mesmas oportunidades de sucesso.
